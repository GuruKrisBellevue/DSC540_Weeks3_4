## Activity 7: Reading Tabular data from a web page and creating Dataframes

##### 1. Read the wikipedia page in a chrome bowser tab and understand the structure

##### 2. Read the page using bs4

```python
# Importing Beautiful soup
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Reading the static GDP file from local drive
path=os.getcwd()
wiki_url=path+"\\List of countries by GDP (nominal) - Wikipedia.htm"
# Using beautiful soup to read the static page contents into soup1 object
with open(wiki_url,'r',encoding="utf-8") as wiki_data:
    soup1=BeautifulSoup(wiki_data)
    print(type(soup1))
```

##### 3. Find the table structure you will need to deal with (how many tables are there)

```python
# Finding the number of tables in the static GDP page using find_all command 
wiki_tables=soup1.find_all('table')
print(f"There are {len(wiki_tables)} tables in the GDP page")
```

##### 4. Find the right table using bs4

```python
#Using class attribute to identify the required table. We use find_all command to identify the tables
# Then parse through each table and get the class attribute
print(f"There are {len(wiki_tables)} tables in the GDP page and below are their classes:")
for table in soup1.find_all('table'):
    print(table.get('class'))

```

```python
# The desired class is the wikitable, hence using that. Using find command to identify the right table
wiki_table1=soup1.find('table',{'class':'"wikitable"|}'})
print(f"The desired table is : \n {wiki_table1}")
```

##### 5. Seperate the source names and the corresponding data

##### ***Approach:*** The HTML page is structured in such a way  that within body section of the main wikitable , the first  table row 'tr' element represents the sources for the HTML page and the second tr element includes the data of subtables in the page. Hence to obtain the source names, we have to parse the tr elements and extract the first occurance . Each source is represented by table data 'td' element.

```python
# Obtaining the list of sources by following the approach explained above by identifying the 'tr' 
# elements and extracting the first occurance
tr_src_elements=wiki_table1.tbody.find_all('tr',recursive=False)[0]
# In the source section, each source element is represented by td element, hence extracting them in a list
list_sources=[td for td in tr_src_elements.find_all('td')]
# Printing the number of sources
print(f"There are {len(list_sources)} sources in the GDP Web page. They are :\n")
print(list_sources)
```

```python
# Getting the data elements by finding the tr elements as we did above and extracting 
# the second occurance of tr. Each data element is represented by td element within the tr. Reading the 
# td elements in a variable for futher parsing
tr_data_elements=wiki_table1.tbody.find_all('tr',recursive=False)[1].find_all('td',recursive=False)
# Within each td element, the data is in tabble format. Hence all the table elements within the 
# td subelement are parsed and stored in a list
list_data_tables=[elem.find_all('table',recursive=False) for elem in tr_data_elements]
# Printing the results and the data
print(f"There are {len(list_data_tables)} Data Elements in the GDP Web page. They are :\n")
print(list_data_tables)
```

##### 6. Get the source names from the list of sources that you have created

```python
# To get the source names, the Text from the sources that was extracted above will be used.
# The source names are stored within the first anchor tag. Hence extracting the text from it
src_name_list=[elem.find_all('a')[0].getText() for elem in list_sources]
src_name_list
```

##### 7. Seperate the header and data from the data that you have seperated before for the first source only and then create a dataframe using that

##### ***Approach: ***  The data elements are stored in a list created above. The first data element is identified and then the header elements and data elements are extracted out of it. To get the Header, we use the table head 'thead' method and find the tr element within it. The table headers are stored as th elements within each 'th' element, hence extracting that and we then strip extra lines and spaces.
##### Same approach is followed to identify the data within the tables. Here we use tbody instead of thead. Approach is very similar.

```python
# Identifying the first data element from the list of data tables
first_src_data=list_data_tables[0][0]
# Extracting the header by extracting the first tr element within the thead element and then extracting
# the th elements within the tr
first_src_tr_header=first_src_data.thead.find_all('tr',recursive=False)[0].find_all('th',recursive=False)
# The header elements are extracted by using the getText method and the results are stored in a list
# after cleansing
first_src_header=[elem.getText().strip('\n') for elem in first_src_tr_header]
first_src_header

```

```python
# Extracting the data within the tables by extracting the tr element within the tbody element and then extracting
# the td elements within the tr. The first occurance of tr is ignored as it does not have the data in correct format.
first_src_tr_data=[elem for elem in first_src_data.tbody.find_all('tr',recursive=False)[1:]]
# Extracting all the td elements within the tr element extracted above
first_src_td_data=[elem.find_all('td',recursive=False) for elem in first_src_tr_data]
# The source data is extracted by extracting the text from the td elements
first_src_data_elem=[ [elem2.getText().strip() for elem2 in elem1] for elem1 in first_src_td_data]
# Printing the top 10 results
first_src_data_elem[:10]
```

```python
# Creating a dataframe of the data and header extracted avove
first_src_df=pd.DataFrame(first_src_data_elem,columns=first_src_header)
first_src_df.head()
```

##### 8. Repeat the task for other two datasources

##### ***Approach:*** As the solution is very similar to step 7, reusable functions are created to extract the data and the headers from each source.

```python
 def remove_spade(item):
     """
     This function is used to cleanse the data by removing the spade and returns clean data
     """
     # Search for spade symbol in the data and if present, using that as a delimiter and 
     # extracting the second element
     if ("♠") in item:
            item=item.split("♠")[1]
     else:
        pass
     return item
```

```python
def create_src_header(src_nbr):
    """
    This function will extract the header from the data sources and returns the header values.
    Receives the sequence of source as input
    """
    # Identifying the data element from the list of data tables
    src_data=list_data_tables[src_nbr][0]
    # Extracting the header by extracting the first tr element within the thead element and then extracting
    # the th elements within the tr
    src_tr_header=src_data.thead.find_all('tr',recursive=False)[0].find_all('th',recursive=False)
    # The header elements are extracted by using the getText method and the results are stored in a list
    # after cleansing
    src_header=[elem.getText().strip('\n') for elem in src_tr_header]
    return src_header
```

```python
def create_src_data(src_nbr):
    """
    This function will extract the data elements from the data sources and returns the data values.
    Receives the sequence of source as input
    """
    # Identifying the data element from the list of data tables
    src_data=list_data_tables[src_nbr][0]
    # Extracting the data within the tables by extracting the tr element within the tbody element and then extracting
    # the td elements within the tr. The first occurance of tr is ignored as it does not have the data in correct format.
    src_tr_data=[elem for elem in src_data.tbody.find_all('tr',recursive=False)[1:]]
    # Extracting all the td elements within the tr element extracted above
    src_td_data=[elem.find_all('td',recursive=False) for elem in src_tr_data]
    # The source data is extracted by extracting the text from the td elements
    src_data_elem=[ [elem2.getText().strip() for elem2 in elem1] for elem1 in src_td_data]
    return src_data_elem
```

```python
def create_dataframe(src_header,src_data_elem):
    """
    This function will create a dataframe out of the data and the headers extracted.
    Returns cleansed dataframe as output
    """
    # Creating dataframe from the source data and headers that are received as input
    src_df=pd.DataFrame(src_data_elem,columns=src_header)
    # Cleansing the data if they contain spade symbols
    for col in src_df.columns:
        src_df[col]=src_df[col].apply(remove_spade)
    return src_df
```

```python
# For loop is used to parse through each data source and extract header and data elements from
# the functions created above and returns a dataframe as a result
for src_nbr in range(0,len(list_sources)):
    # Exxtracting the Data header
    src_header=create_src_header(src_nbr)
    # Extracting the data elements
    src_data_elem=create_src_data(src_nbr)
    # Creating dataframe 
    src_df=create_dataframe(src_header,src_data_elem)
    # Displaying the results
    print(f"Displaying the Dataframe of Source #{src_nbr+1} ")
    print(src_df.head())
    print("*"*50)
    
```

```python
def check_duplicates(col):
    """
    This function receives column name as input and prints the results if there are any duplicates
    """
    # Extracting the original number of rows in the dataframe
    orig_len=len(visit_df[col])
    # Extracting the number of rows in the dataframe after getting the unique count
    uniq_len=len(visit_df[col].unique())
    # If  there are duplicates, print the number of duplicates
    if orig_len-uniq_len>0:
        print(f"There are duplicates in {col} field")
    else:
        print(f"There are NO duplicates in {col} field")      
```
